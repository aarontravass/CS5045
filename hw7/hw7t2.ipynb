```python
import time
import matplotlib.pyplot as plt
import copy
import numpy as np
from matplotlib.ticker import FormatStrFormatter
```

```python
def insertion_sort(mylist):
    unsorted_list=copy.deepcopy(mylist)
    start = time.time()
    for i in range(1, len(unsorted_list)):
        search_index = i
        insert_value = unsorted_list[i]
        
        while search_index > 0 and unsorted_list[search_index-1] > insert_value:
            unsorted_list[search_index] = unsorted_list[search_index-1]
            search_index -= 1
            
        unsorted_list[search_index] = insert_value
    end = time.time()
    return end-start
```

```python
def another_insert_sort(unsorted):
    mylist=copy.deepcopy(unsorted)
    start = time.time()
    nums=len(mylist)
    for i in range(1,nums):
        for j in range(i-1,-1,-1):
            if(j>0):
                if mylist[j-1]<=mylist[i]<=mylist[j]:
                    mylist.insert(j,mylist.pop(i))
                    break
                elif mylist[i]<=mylist[0]:
                    mylist.insert(0,mylist.pop(i))
                    break
            else:
                if mylist[i]<mylist[0]:
                    mylist.insert(0,mylist.pop(i))
    end = time.time()
    return end-start
```

```python
a=[]
b=[]
x_len=10
barWidth = 0.25
fig,ax = plt.subplots(figsize =(8, 12))
ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(0, 10, 0.25))
for __ in range(x_len):
    mylist=list(np.random.randint(low = 1,high=999,size=10000))
    a.append(insertion_sort(mylist))
    b.append(another_insert_sort(mylist))
plt.bar([i for i in range(1,1+x_len)],a,color='b',width = barWidth,label="insertion sort fn")
plt.bar([i+barWidth for i in range(1,1+x_len)],b,color='g',width = barWidth,label="another insertion sort fn")
plt.xlabel('Runs', fontweight ='bold', fontsize = 15)
plt.ylabel('Times in seconds', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth/2 for r in range(1,x_len+1)], list(range(1,x_len+1)))
plt.legend()
plt.show()
```
